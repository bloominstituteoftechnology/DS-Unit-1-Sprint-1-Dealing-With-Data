```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('LoanStats_2018Q3.csv', skiprows=1, skipfooter=2)
```

```python
test_string = '17.97%'
def remove_percent(string):
    return float(string.strip('%'))
remove_percent(test_string)
```

```python
df['int_rate'] = df['int_rate'].apply(remove_percent)
```

```python
df['int_rate'].head()
```

```python
df['emp_title'].value_counts().head(20)
```

```python
df['emp_title'].isnull().sum()
```

```python
examples = ['owner', 'Supervisor ',
            'Project manager', np.nan]
def clean_title(x):
    if isinstance(x, str):
        return x.strip().title()
    else:
        return 'Unknown'

for example in examples:
    print (clean_title(example))
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
```

```python
df['emp_title'].value_counts().head(20)
```

```python
df['emp_title_manager'] = df['emp_title'].str.contains('Manager')
```

```python
df['emp_title_manager'].value_counts()
```

```python
df['issue_d'] = pd.to_datetime(df['issue_d'],
                              infer_datetime_format=True)
df['issue_d'].describe()
```

```python
df['issue_year'] = df['issue_d'].dt.year
df['issue_month'] = df['issue_d'].dt.month
df['issue_month'].sample(n=10).values
```

```python
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'],
                                       infer_datetime_format=True)
```

```python
(df['issue_d'] - df['earliest_cr_line']).head()
```

```python
df['days_from_earliest_credit_to_issue'] = (
    df['issue_d'] - df['earliest_cr_line']).dt.days
```

```python
df['days_from_earliest_credit_to_issue'].head().values
```

```python
[col for col in df if col.endswith('_d')]
```

```python
for col in ['issue_d', 'last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d']:
    df[col] = pd.to_datetime(
        df[col],
        infer_datetime_format=True)
```

```python
df['term'].value_counts(dropna=False)
```

```python
df['term'] = df['term'].str.strip('months').astype(int)
```

```python
df['term'].value_counts(dropna=False)
```

```python
df['loan_status'].value_counts(dropna=False)
```

```python
def great(loan_status):
    if loan_status == 'Current' or loan_status =='Fully Paid':
        return 1
    else:
        return 0
    
df['loan_status_is_great'] = df['loan_status'].apply(great)
```

```python
df['loan_status_is_great'].value_counts(dropna=False)
```

```python
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year
```

```python
# I am going to guess 'revol_util' shows what % of revolving credit is 
# in use right now
df['revol_util'].head()
# looks right
```

```python
#df['revol_util'] = df['revol_util'].apply(remove_percent)
df['revol_util'].dtypes
df['revol_util'].isnull().sum()
```

```python
#129 null values. I'll need to fill them.
df['revol_util'].shape
print('Print the df before fillna and strip %:',df['revol_util'])
df['revol_util'] = df['revol_util'].fillna('0').str.strip('%').astype(float)
print('Print the df after fillna and strip %:',df['revol_util'])
```

```python
df['emp_title'].nunique()
```

```python
common20 = df['emp_title'].value_counts().head(20).index
```

```python
def title_sorter(title):
  if title in common20:
    return title
  else:
    return 'Other'
df['emp_title'] = df['emp_title'].apply(title_sorter)
```
