```
import numpy as np
import pandas as pd 
```

```
df = pd.read_csv("https://projects.fivethirtyeight.com/trump-approval-data/approval_polllist.csv")
print(df.shape)
```

```
df.head()
```

```
df.isnull()
```

```
date = df.startdate
yes = df.approve
no = df.disapprove
```

```
print(date)
```

```
import matplotlib.pyplot as plt
```

```
df.dtypes
```

```
plt.scatter(date, yes)
plt.title('Yup')
```

```
plt.scatter(date, no)
plt.title('Nope')
```

```
df["unsure_delta"] = df.approve + df.disapprove
df["unsure"] = 100 - df.unsure_delta
maybe = df.unsure
```

```
plt.scatter(date, maybe, s=4)
plt.title('Well, you know')
```
