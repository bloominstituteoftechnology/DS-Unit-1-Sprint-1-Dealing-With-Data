```
def confidence_interval(data, alpha=.9):
  n = len(data)
  mean = sum(data)/n
  data = np.array(data)
  stderr = stats.sem(data)
  t = stats.t.ppf((1 + alpha) / 2.0, n-1)
  interval = stderr * t
  return (mean , mean-interval, mean+interval)
```
