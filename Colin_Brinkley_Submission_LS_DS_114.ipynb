```python
#Import the CSV and trim off the unneeded rows
import pandas as pd
loanDF = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2)
pd.set_option('display.max_columns',100)
loanDF.head()
```

```python
#Checking the size of the DF for later reference
loanDF.shape
```

```python
#Checking which columns are empty
loanDF.isnull().sum().sort_values(ascending=False)
```

```python
#Drop the empty columns
loanDF = loanDF.drop(columns=['id', 'member_id', 'desc', 'url'])
```

```python
#Check the formatting of data
loanDF.head()
```

```python
#Drop the percent signs and convert the numbers to floats
def strip_percent_str(string):
  return float(string.strip('%'))

loanDF['int_rate'] = loanDF['int_rate'].apply(strip_percent_str)
```

```python
#Make sure everything looks good
loanDF['int_rate'].head(10)
```

```python
#Check professions
loanDF['emp_title'].value_counts(dropna=False).head(20)
```

```python
#Remove blank spaces from profession titles, make them uppercase, and replace the NaN values with "Unknown Profession"
def clean_spaces(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return "Unknown Profession"

loanDF['emp_title'] = loanDF['emp_title'].apply(clean_spaces)
```

```python
#Make a Manager column
loanDF['emp_title_manager'] = True
loanDF['emp_title_manager'] = loanDF['emp_title'].str.contains('Manager')
loanDF['emp_title_manager']
```

```python
#Make a column about having a great loan status and checked if everyone's loan status is really that great
loanDF['loan_status_is_great'] = 0

for _ in range(loanDF.shape[0]):
  if (loanDF.at[_, 'loan_status'] == 'Current') or (loanDF.at[_, 'loan_status'] == 'Fully Paid'):
    loanDF.at[_, 'loan_status_is_great'] = 1
```

```python
#Checking out everyone's great loan statuses
loanDF['loan_status_is_great'].head(100)
```

```python
#Convert 'last_pymnt_d' to the Date/Time format
loanDF['last_pymnt_d'] = pd.to_datetime(loanDF['last_pymnt_d'], infer_datetime_format=True)
```

```python
#Create "last_pymnt_d_month" and "last_pymnt_d_year" columns
loanDF['last_pymnt_d_month'] = loanDF['last_pymnt_d'].dt.month
loanDF['last_pymnt_d_year'] = loanDF['last_pymnt_d'].dt.year
```
