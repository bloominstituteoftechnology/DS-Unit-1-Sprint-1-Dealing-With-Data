### Convert `int_rate`

Define a function to remove percent signs from strings and convert to floats

```python
def remove_percent(string):
  return float(string.strip('%'))
```

Apply the function to the `int_rate` column

```python
df['int_rate'] = df['int_rate'].apply(remove_percent)
```

### Clean `emp_title`

Look at top 20 titles

```python
df['emp_title']
```

```python
print(df['emp_title'].value_counts(dropna=False).shape)
df['emp_title'].value_counts(dropna=False).head(20)
```

How often is `emp_title` null?

```python
df['emp_title'].isnull().sum()
```

Clean the title and handle missing values

```python
import numpy as np
examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]
```

```python
def clean_title(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return "Unknown"
  return title.strip()

[clean_title(x) for x in examples]
```

```python
# df = df.apply(lambda x: x.strip().title() if isinstance(x, str))
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
print(df['emp_title'].value_counts(dropna=False).shape)
df['emp_title'].value_counts(dropna=False).head(20)
```

```python
def remove_percent(string):
  return float(string.strip('%'))
```

```python
df2 = df.copy()
df2['int_rate'] = df2['int_rate'].apply(remove_percent)
df2.head()
```

```python
df3 = df2.copy()
import numpy as np
def change_titles(titlestring):
  if isinstance(titlestring, str):
    return titlestring.strip().title()
  else:
    return "Unknown"
  return titlestring.strip()

df3['emp_title'] = df3['emp_title'].apply(change_titles)
df3['emp_title'].value_counts().head(10)
```

```python
df3['emp_title'].value_counts(dropna=False).head(10)
```

```python
df4 = df3.copy()
df4['is_a_manager'] = df4['emp_title'].str.contains('Manager', case=False)
# df4.head(10)
df4[df4['is_a_manager']==True].head(20)
```

# Convert the `term` column from string to integer.


```python
df5 = df4.copy()

#function to strip whitespaces then the string "months" and finally cast the result as an int
def convert_to_int(string):
  string = string.strip()
  string = string.strip('months')
  return int(string)

df5['term'] = df5['term'].apply(convert_to_int)
df5['term'].dtypes
```

# Make a column named `loan_status_is_great`. It should contain the integer 1 if `loan_status` is "Current" or "Fully Paid." Else it should contain the integer 0.


```python
# cell just for reviewing and comparing the data
df5['loan_status'].head(30)
```

```python
df6 = df5.copy()

#convert boolean to int
def bool_to_int(value):
  return int(value)

df6['loan_status_is_great'] = df6['loan_status'].str.contains('Current|Fully Paid') # '|' acts as an OR
df6['loan_status_is_great'] = df6['loan_status_is_great'].apply(bool_to_int)
df6[['loan_status','loan_status_is_great']].head(30)
```

# Make `last_pymnt_d_month` and `last_pymnt_d_year` columns.

```python
df7 = df6.copy()
df7['last_pymnt_d'].isnull().sum()
```

```python
# convert the columns to date values in-line and get the month or year as needed
# if the values need to be int then make a function and .apply it to those columns as well as dealing with NaN values
df7['last_pymnt_d_month'] = pd.to_datetime(df7['last_pymnt_d'],infer_datetime_format=True).dt.month
df7['last_pymnt_d_year'] = pd.to_datetime(df7['last_pymnt_d'],infer_datetime_format=True).dt.year
df7.head()
```
