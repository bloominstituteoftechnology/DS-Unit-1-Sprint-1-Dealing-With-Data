```python
#find out if there is a way to shorten my test and function with a dictionary.**

#return 1 if string in ["Current" , "Fully Paid"] else 0 **

def assign_loan_status(string):
    if string == 'Current':
        return 1

    elif string == 'Fully Paid':
        return 1
    else:
        return 0
```
