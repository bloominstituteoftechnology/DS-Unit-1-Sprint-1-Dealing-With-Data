# Crosstab Review

```python
import pandas as pd

df = pd.DataFrame({'animal': ['cat', 'cat', 'dog', 'cat', 'dog', 'cat', 'dog', 'dog'], 
                   'country': ['USA', 'Canada', 'Canada', 'Canada', 'USA', 'Canada','USA', 'USA'], 
                   'gender': ['male','female','male','male','male','male','female','female']})
df.head()
```

```python
pd.crosstab(df['animal'], df['country'])
# First argument becomes row index
```

```python
pd.crosstab(df['animal'],df['gender'], normalize='columns')
```

```python
pd.crosstab(df['animal'],df['country'], normalize='index')
# for more than two variables, add a list of multiple variables on the same dimension
```

```python
crosstab = pd.crosstab(df['country'], [df['gender'],df['animal']])
pd.crosstab(df['country'], [df['gender'],df['animal']])
```

```python
crosstab.plot()
# Crosstabs are dataframes like any other
```

```python
pd.crosstab(df['country'], [df['gender'],df['animal']], margins=True)
```

```python
pd.crosstab([df['country'],df['gender']],df['animal'])
```

### Convert `int_rate`

Define a function to remove percent signs from strings and convert to floats

```python
df['int_rate']
```

```python
int_rate = '15.02%'
# What can we do with this? Whatever we do, we can do with every cell, because series are iterable
int_rate[:-1]
# Just like a list comprehension
```

```python
int_list = ['15.02%', '13.56%', '16.91']
```

```python
int_list[:2]
```

Apply the function to the `int_rate` column

```python
float(int_rate.strip('%'))
```

```python
type(float(int_rate.strip('%')))
```

```python
def remove_percent_to_float(string):
  return float(string.strip('%'))
```

```python
[remove_percent_to_float(item) for item in int_list]
```

```python
# .apply applies a function to an entire iterable
df['int_rate'] = df['int_rate'].apply(remove_percent_to_float)
```

```python
df['int_rate']
```

### Clean `emp_title`

Look at top 20 titles

```python
df['emp_title'].value_counts(dropna=False).head(20)
```

How often is `emp_title` null?

```python
df['emp_title'].isnull().sum()
```

```python
import numpy as np
type(np.NaN)
```

Clean the title and handle missing values

```python
examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]
def clean_title(item):
  if isinstance(item, str):
    return item.strip().title()
  else:
    return 'Unknown'
[clean_title(item) for item in examples]
```

```python
df['emp_title'] = df ['emp_title'].apply(clean_title)
df.head()
```

```python
df['emp_title'].value_counts(dropna=False).head(20)
```

```python
df['emp_title'].value_counts(dropna=False).reset_index().shape
# Unique job titles
```
