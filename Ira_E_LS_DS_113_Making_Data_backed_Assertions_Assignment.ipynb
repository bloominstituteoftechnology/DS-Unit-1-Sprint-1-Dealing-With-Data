```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

persons_url = 'https://raw.githubusercontent.com/pragmatizt/DS-Unit-1-Sprint-1-Dealing-With-Data/master/module3-databackedassertions/persons.csv'
df = pd.read_csv(persons_url)
```

```
print(df.head())
df.shape
```

```
# Let's put it on DataFrame for easier viewing.

persons = pd.DataFrame(df)
persons.head()
```

```
# Let's play around with a crosstab

pd.crosstab(persons['age'], persons['exercise_time'])

# shows a result that's a bit crazy.  Let's bin both age, and exercise time.
```

```
# start with the age bins

age_bins = pd.cut(persons['age'], 3) # 4 equal sized bins
exercise_bins = pd.cut(persons['exercise_time'], 3) # 4 equal sized bins as well
weight_bins = pd.cut(persons['weight'], 3)
```

```
# Now let's do a crosstab with both bins.

pd.crosstab(age_bins, exercise_bins)
```

```
# Normalized crosstab of the three variables
three_variables_norm = pd.crosstab(exercise_bins, [age_bins, weight_bins], normalize='columns')
three_variables_norm.head()
```

```
# Non-normalized
three_variables = pd.crosstab(exercise_bins, [age_bins, weight_bins])
three_variables.head()
```

```
# Well this plot graph tells me nothing.
three_variables.plot(legend=False);
```

```
# Also nothing.
three_variables.plot(kind='bar', legend=False);
```

```
# Let's try a stacked bar chart.
three_variables.plot(kind='bar', stacked=True, legend=False);

# I lowered the bins from 5 to 3 to simplify the graphs.  But still, they don't really tell a story by themselves. 
```

```
# 3D Scatterplot on matplotlib
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

xs = persons['exercise_time']
ys = persons['age']
zs = persons['weight']
ax.scatter(xs, ys, zs, s=50, alpha=0.6, edgecolors='w')

ax.set_xlabel('Exercise Time')
ax.set_ylabel('Age')
ax.set_zlabel('Weight')

plt.show()
```

```
# Let's see if seaborn yields anything.  
# Relationship between weight and exercise time.
sns.jointplot(x='exercise_time', y='weight', data=persons, kind='reg');
```

```
# Relationship between age and weight

sns.jointplot(x='age', y='weight', data=persons, kind='reg');
```

```
sns.jointplot(x='age', y='exercise_time', data=persons, kind='reg');
```
