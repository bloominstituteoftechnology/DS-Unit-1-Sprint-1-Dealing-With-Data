## Part 1 - Load and validate the data

```
# TODO

#LOAD DATA
cancer_survival_url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/haberman/haberman.data'
```

```
#READ DATA
cancer_survival = pd.read_csv(cancer_survival_url)
```

```
#VALIDATE NUMBER OF OBSERVATIONS
df= pd.DataFrame(cancer_survival)
df.head(5)
```

```
#VALIDATE MISSING VALUES
df.isnull()
```

```
# ADD APPROPRIATE NAMES FOR COLUMNS
column_headers = ["Age of patient at time of operation", "Patient's year of operation",
                  "Number of positive axillary nodes detected", "Survival status"]
df = pd.read_csv(cancer_survival_url, names=column_headers)
df.head(5)
```

```
#CHANGE TWO TO ZERO FOR NOT SURVIVAL
def two_to_zero(n):
  if n == 2:
    return 0
  else:
    return 1
df['Survival status'] = df['Survival status'].apply(two_to_zero)
df
```

## Part 3 - DataFrame Filtering

Use DataFrame filtering to subset the data into two smaller dataframes. You should make one dataframe for individuals who survived >5 years and a second dataframe for individuals who did not. 

Create a graph with each of the dataframes (can be the same graph type) to show the differences in Age and Number of Positive Axillary Nodes Detected between the two groups.

```
# TODO
#Subset survivor
survivor=df[df['Survival status'] ==1]
```

```
#Subset survivor vs Age and Node
age_bins_1=pd.cut(survivor['Age of patient at time of operation'],4)
node_bins = pd.cut(survivor['Number of positive axillary nodes detected'],4)
survivor_sub=pd.crosstab(survivor['Survival status'],[age_bins_1,node_bins])
survivor_sub
```

```
#Survivor graph
survivor_sub.plot(kind='bar').legend(bbox_to_anchor=(1.6, 1.0))
```

```
#Subset non-survivor
non_survivor=df[df['Survival status'] ==0]
non_survivor
```

```
#Subset non-survivor vs age and node
age_bins_2 = pd.cut(non_survivor['Age of patient at time of operation'],4)
node_bins_2 = pd.cut(non_survivor['Number of positive axillary nodes detected'],4)
non_survivor_sub= pd.crosstab(non_survivor['Survival status'],[age_bins_2,node_bins_2])
non_survivor_sub
```

```
#Non_survivor graph
non_survivor_sub.plot(kind='bar').legend(bbox_to_anchor=(1.6,1.0))
```
