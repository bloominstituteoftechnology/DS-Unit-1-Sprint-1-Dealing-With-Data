```python
plt.style.use('fivethirtyeight')

fake = pd.Series([38, 3, 2, 1, 2, 4, 6, 5, 5, 33], index=range(1,11))

#PLOT THE BAR GRAPH
ax = fake.plot.bar(color='#EC713B', width=0.9)

#LABEL X AND Y AXIS, AND CUSTOMIZE THE TICKS FOR Y-AXIS
ax.set(xlabel='Rating', 
       ylabel='Percent of total votes',
       yticks=range(0,50,10))

#ADD TEXT AND MODIFY THE LOCATION OF THIS TEXT VIA THE X & Y AXIS
ax.text(x=-2, y=50, s="'An Inconvenient Sequel: Truth to Power' is divisive", 
        fontsize=16, fontweight='bold')

#ADD ANOTHER LINE OF TEXT REPRESENTING THE SUBTITLE
ax.text(x=-2, y=46, s="IMDb ratings for the film as of Aug. 29")

ax.tick_params(labelrotation=0) #ROTATED THE X-AXIS NUMBERS REPRESENTING VALUES
```
