# Tips for choosing an appropriate visualization:

## Use Appropriate "Visual Vocabulary"

[Visual Vocabulary - Vega Edition](http://ft.com/vocabulary)

## What are the properties of your data?
- Is your primary variable of interest continuous or discrete?
- Is in wide or long (tidy) format?
- Does your visualization involve multiple variables?
- How many dimensions do you need to include on your plot?

Can you express the main idea of your visualization in a single sentence?

How hard does your visualization make the user work in order to draw the intended conclusion?

## Annotate with text

```python
plt.style.use('fivethirtyeight')
fig = plt.figure()
fake = pd.Series([38, 3, 2, 1, 2, 4, 6, 5, 5, 33], 
                 index=range(1, 11))

ax = fake.plot.bar(color='#ef7030', width=0.9)

ax.set(yticks=range(0, 50, 10), facecolor='#f9f9f9')
fig.patch.set_facecolor('#f9f9f9')

plt.ylabel('Percent of total votes', fontsize=9, fontweight='bold')
plt.xlabel('Rating', fontsize=9, fontweight='bold')

ax.text(x=-2, y=46, s="'An Inconvenient Sequel: Truth To Power' is divisive",
        fontsize=12, fontweight='bold')
ax.text(x=-2, y=43, s='IMDb ratings for the film as of Aug. 29',
        fontsize=11)
ax.tick_params(labelrotation=0)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv')

df.head()
```

```python
# We will filter the data to capture the major that interest us.  20% in 1970

df_filtered = df.loc[0, df.loc[0] < 20]
df_filtered
```

```python
# put the the name of the majors in a list
df_filtered = pd.DataFrame(df_filtered)
df_filtered
```

```python
majores = list(df_filtered.index.values) 
```

```python
majores
```

```python
ax = df.plot(x = 'Year', y = majores,figsize = (12,8))
```

```python
# Setting the style from matplotlib to '538'
plt.style.use('fivethirtyeight')

# List a list of colors
colors = [[0,0,0], [230/255,159/250,0], [86/255,180/255,233/255], [0,158/255,115/255], [213/255,94/255,0], [0,114/255,178/255]]

# Plot and define - y is from the list of majores under 20% in 1970 - color is taken from the list of colors.
ax = df.plot(x = 'Year', y = majores, figsize = (12,8), color = colors, legend= False)

# Set the size of the number of values and the x and y ticks
ax.tick_params(axis = 'both', labelsize = 18)

# Set the y labels, although the y lables are set automatically we have edited in a string format so that we can the % and added space after the numbers so that they are aligned. the number 10 is to set the difference between each label
ax.set_yticklabels(labels = [ 10,'0   ', '10   ', '20   ', '30   ', '40   ', '50%'])


# Set a black horizontal line, set it on the x axis (y=0) make it bold
ax.axhline(y = 0, color = 'black',linewidth = 2, alpha = 1)

# Add limit of x axis so that we can add the small ticks on the y axis
ax.set_xlim(left = 1969, right = 2011)

# Remove the X label (year)
ax.xaxis.label.set_visible(False)

# Add text:

# Add the signature mark, set the x and y, background color and font color.

ax.text(x = 1966, y = -7, s = ' DATAQUEST                    Source: National Center for Education Statistics - Replicated for Lambda School', fontsize = 15, color= '#f0f0f0', backgroundcolor = 'grey')

# Title
ax.text(x = 1966.5, y = 60, s = 'The gender gap is transitory - even for extreme cases', fontsize = 20, color= 'black', alpha = 0.8,fontweight ='bold' )
ax.text(x = 1966.5, y = 55, s = 'Percentage of Bachelors conferred to women from 1970 to 2011 in the US for\nextreme cases where the percentage was less than 20% in 1970', fontsize = 16, color= 'black', alpha = 0.9)

# We remove the legend and add these texts to the lines, we add them based on where we want thme to be by defining their x and y, and rotation. colors is indexed from the colors list. Before we turn the legend  - False we make sure they are correct.
# and we use rotation to play with the text's alignment
ax.text(x = 1994, y = 44, s = 'Agriculture', color = colors[0], weight = 'bold', rotation = 33,
              backgroundcolor = '#f0f0f0')
ax.text(x = 1985, y = 44, s = 'Architecture', color = colors[1], weight = 'bold', rotation = 25,
              backgroundcolor = '#f0f0f0')
ax.text(x = 1979, y = 43, s = 'Business', color = colors[2], weight = 'bold', rotation = 50,
              backgroundcolor = '#f0f0f0')
ax.text(x = 2001, y = 29.5, s = 'Computer Science', color = colors[3], weight = 'bold', rotation = -34,
              backgroundcolor = '#f0f0f0')
ax.text(x = 1985, y = 16, s = 'Engineering', color = colors[4], weight = 'bold', rotation = 5,
              backgroundcolor = '#f0f0f0')
ax.text(x = 1977, y = 25, s = 'Physical Sciences', color = colors[5], weight = 'bold', rotation = 25,
              backgroundcolor = '#f0f0f0')



```
