```
# Created by Andrew Mikol and Zachery Moy

import pandas as pd
```

```
df = pd.read_csv('https://raw.githubusercontent.com/mikvikpik/Tuesday-June-Fourth/master/survey.csv',
                names = ['name', 'prog_level', 'math_level', 'finish_level', 'classroom'],
                skiprows=1)
df.head()
```

```
# At first glance we have a small amount of data with 30 observations and 5 attributes.
# 3 of the 5 attributes are numeric in value
# Name is indeterminate
# In a classroom is a binary variable
# There is one observation that is Null across the board and has no valuable information
df.shape
```

```
df.head(30)
```

```
df.describe()
# Most people feel less confident in their programming by mean
# Most people feel fairly confident in their math level by mean
# Most people want to be extremely confident by the end of LS
```

```
df['prog_level']= df['prog_level'].fillna(value=3)
```

```
df['math_level'] = df['math_level'].fillna(value=6)
```

```
df.head(30)
```

```
df['prog_level'] = df['prog_level'].astype(int)
df['math_level'] = df['math_level'].astype(int)
```

```
df.plot.scatter(df.prog_level, df.math_level)
```
